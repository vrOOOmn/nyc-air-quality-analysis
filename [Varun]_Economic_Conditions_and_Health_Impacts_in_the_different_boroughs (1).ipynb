# Dot Plot showing Neighborhood Poverty with respect Asthma ER Visits due to PM2.5 (By Borough)

```python
df = pd.merge(new_health_impacts2, new_economic_conditions5, on='Geography')
df = df.loc[:, ~df.columns.str.lower().str.contains('geotype|geoid|timeperiod')]
new_labels = {
    'Estimated annual rate (age 18+) per 100,000 adults': 'Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (age 18+)',
    'Estimated annual rate (under age 18) per 100,000 children': 'Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (under age 18)',
    'Percent': 'Neighborhood Poverty (%)'
}
df = df.rename(columns = new_labels)
grouped_by_borough = df.groupby('Geography').mean().round(3).reset_index()


# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 6))


# Plot 1: Adult Rate of Asthma ER visits due to PM2.5 Rate of  Neighborhood Poverty %
axes[0].scatter(grouped_by_borough['Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (age 18+)'],
                grouped_by_borough['Neighborhood Poverty (%)'], s=100)
for i, borough in enumerate(grouped_by_borough['Geography']):
    axes[0].annotate(borough,
                    (grouped_by_borough['Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (age 18+)'][i],
                     grouped_by_borough['Neighborhood Poverty (%)'][i]),
                    xytext=(5, 5), textcoords='offset points')
axes[0].set_xlabel('Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (age 18+)')
axes[0].set_ylabel('Neighborhood Poverty (%)')
axes[0].set_title('Asthma ER visits due to PM2.5 (age 18+) vs Neighborhood Poverty (%)')

# Plot 2: Adult Rate of Asthma ER visits due to PM2.5 Rate of  Neighborhood Poverty %
axes[1].scatter(grouped_by_borough['Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (under age 18)'],
                grouped_by_borough['Neighborhood Poverty (%)'], s=100)
for i, borough in enumerate(grouped_by_borough['Geography']):
    axes[1].annotate(borough,
                    (grouped_by_borough['Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (under age 18)'][i],
                     grouped_by_borough['Neighborhood Poverty (%)'][i]),
                    xytext=(5, 5), textcoords='offset points')
axes[1].set_xlabel('Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (under age 18)')
axes[1].set_ylabel('Neighborhood Poverty (%)')
axes[1].set_title('Asthma ER visits due to PM2.5 (under age 18) vs Neighborhood Poverty (%)')

plt.tight_layout()


# Print the merged data
print("\nBorough Data:")
print(grouped_by_borough.to_string(index=False))
print(grouped_by_borough[['Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (age 18+)',
                          'Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (under age 18)',
                          'Neighborhood Poverty (%)']].median())

plt.show()
```

# Correlation Matrix between Neighborhood Poverty and Asthma ER Visits due to PM2.5

```python

df_copy = df.drop('Geography', axis = 1).copy()

correlation_matrix = df_copy.corr()


correlation_matrix.rename(index=new_labels, inplace=True)


# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.3f', linewidths=0.5)

# Add title for clarity
plt.title('Correlation Matrix', fontsize=16)

# Display the plot
plt.show()
```

# Confidence Intervals for the correlations

```python
from scipy.stats import pearsonr, norm


col1 = df_copy["Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (age 18+)"]
col2 = df_copy["Average Annual Rate (per 100k) of Asthma ER visits due to PM2.5 (under age 18)"]
col3 = df_copy["Neighborhood Poverty (%)"]

# Function to calculate correlation and confidence interval
def correlation_analysis(x, y, confidence=0.95):
    r, _ = pearsonr(x, y)  # Calculate Pearson correlation
    n = len(x)  # Number of samples
    z = np.arctanh(r)  # Fisher transformation
    se = 1 / np.sqrt(n - 3)  # Standard error
    z_critical = norm.ppf(1 - (1 - confidence) / 2)  # Z critical value for the confidence level
    lower = np.tanh(z - z_critical * se)  # Transform back to correlation
    upper = np.tanh(z + z_critical * se)
    return r, lower, upper

# Define pairs for analysis
pairs = [
    (col1, col3, "Asthma ER visits (adults) vs. Poverty"),
    (col2, col3, "Asthma ER visits (children) vs. Poverty"),
    (col1, col2, "Asthma ER visits (adults) vs. Asthma ER visits (children)")
]

# Step 5: Perform analysis
results = {
    label: {
        "Correlation": r,
        "CI Lower": ci_lower,
        "CI Upper": ci_upper
    }
    for x, y, label in pairs
    for r, ci_lower, ci_upper in [correlation_analysis(x, y)]
}

for pair, stats in results.items():
    print(f"{pair}:")
    print(f"  Correlation: {stats['Correlation']:.3f}")
    print(f"  95% Confidence Interval: [{stats['CI Lower']:.3f}, {stats['CI Upper']:.3f}]")
```
